# Importing required libraries

```python
import seaborn as sns
import matplotlib.pyplot as plt 
import bar_chart_race as bcr 
import pandas as pd 
import geopandas as gpd
import geoplot as gplt
import geoplot.crs as gcrs
import warnings 
warnings.filterwarnings('ignore', '.*do not.*', )

df=pd.read_csv("final.csv")
```

# Having a glance at the raw data set

```python
df.head()
```

# Removing unneccessary words

```python
df.columns=df.iloc[0]
df.drop([0],axis=0,inplace=True)
```

# Dropping unrequired columns

```python
dff=df.drop(columns=['Sl.\nNo.','Total A','Total S'])
dff.head()
```

# Suicidal Deaths Subset

```python
con=['Year','Male S','Female S']
subset=dff[con]

subset.head()
```

# Data Preprocessing

```python
subset['Male S'] = subset['Male S'].astype('int')
subset['Female S'] = subset['Female S'].astype('int')
```

# Plotting Line Chart for Male vs Female Suicides (1967 - 2019)

```python
x=subset['Year'].to_numpy()
y=subset['Male S'].to_numpy()
z=subset['Female S'].to_numpy()
plt.figure(figsize=(25,10))
plt.plot(x,y)
font = {'family': 'serif',
        'color':  'darkred',
        'weight': 'normal',
        'size': 20,
        }
plt.title('Plotting Line Chart for Male vs Female Suicides (1967 - 2019)',fontdict = font)
plt.xlabel("Year",fontdict=font)
plt.ylabel("Male and Female Suicides",fontdict=font)
plt.plot(x,z)
plt.legend(['Male','Female'],loc = 'best')
plt.xticks(rotation = 90)
plt.show()
```

# Plotting Bar Chart for Male vs Female Suicides (1967 - 2019)

```python
font = {'family': 'serif',
        'color':  'darkred',
        'weight': 'normal',
        'size': 20,
        }
plt.figure(figsize=(25,10))
plt.bar(x, y, color = "red")
plt.xlabel("Year",fontdict = font)
plt.ylabel("Male suicides",fontdict = font)
plt.xticks(rotation = 90)
plt.title('Plotting Bar Chart for Male vs Female Suicides (1967 - 2019)',fontdict = font)
plt.bar(x,z,color='green')
plt.xlabel("Year",fontdict = font)
plt.ylabel("Female suicides",fontdict = font)
plt.xticks(rotation = 90,fontsize = 16)
plt.yticks(fontsize = 16)
plt.legend(['Male','Female'],loc = 2,fontsize = 16)
plt.title('Plotting Bar Chart for Male vs Female Suicides (1967 - 2019)',fontdict = font)
plt.show()
```

# Male Suicide Percentage for First and Last 10 years of selected dataset

```python
p1=subset.iloc[0:11]
p2=subset.iloc[42:53]
font = {'family': 'serif',
        'color':  'darkred',
        'weight': 'normal',
        'size': 20,
        }
plt.figure(figsize=(15,12))
sns.set_theme(style="ticks",font_scale=0.6)
plt.subplots_adjust(wspace=0.2,hspace=0.6)
plt.subplot(1,2,1)
plt.title('Suicides in 1967 - 1976',fontdict = font)
plt.pie(x=p1['Male S'], labels=p1['Year'], autopct='%1.1f%%', shadow=True, startangle=90,textprops={'size': 'x-large'})
plt.subplot(1,2,2)
plt.title('Suicides in 2009 - 2019',fontdict = font)
plt.pie(x=p2['Male S'], labels=p2['Year'], autopct='%1.1f%%', shadow=True, startangle=90,textprops={'size': 'x-large'})
plt.show()
```

# Running Bar Chart showing Alarming Suicide Rates (Male vs Female)

```python
subset.set_index('Year',inplace=True)
csum=subset.cumsum(axis=0)  # gettg cumulative sumin
csum.head()
```

```python
bcr.bar_chart_race(df=csum,filename=None ,figsize=(5,5),title=" Suicides in Male vs Female over the years")
```

# Plotting Bar Chart (Accidental Deaths)

```python
font = {'family': 'serif',
        'color':  'darkred',
        'weight': 'normal',
        'size': 20,
        }
plt.figure(figsize=(25,10))
plt.bar(x, y, color = "red")
plt.title('Male vs Female Accidental Deaths',fontdict = font)
plt.xlabel("Year",fontdict = font)
plt.xticks(rotation = 90,fontsize = 16)
plt.ylabel("Male accidental deaths",fontdict = font)
plt.yticks(fontsize = 16)
plt.bar(x,z,color='green')
plt.xlabel("Year",fontdict = font)
plt.ylabel("Female accidental deaths",fontdict = font)
plt.legend(['Male','Female'],loc = 2,fontsize = 16)
plt.show()
```
